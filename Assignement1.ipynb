## 2. Student Scores Distribution
Analyze the distribution of scores for a class of 100 students in two subjects: Math and Science.

Your Task:

NumPy: Generate two arrays of 100 random scores each. math_scores should follow a normal distribution with a mean of 75 and a standard deviation of 10. science_scores should have a mean of 80 and a standard deviation of 15.

Pandas: Create a DataFrame from these NumPy arrays.

Matplotlib: Create a figure with two subplots:

The first subplot should be an overlapping histogram showing the distribution of scores for both Math and Science. Use transparency (alpha) to make both distributions visible.

The second subplot should be a scatter plot comparing math_scores (x-axis) against science_scores (y-axis) to see if there's a correlation

```python
import numpy as np
import pandas as pd

```

```python
#mathscores shall have mean of 75 and sd of 10
math_scores=np.random.normal(loc=75,scale=10,size=100)

#science score shall have mean of 80 and sd of 15
sci_scores=np.random.normal(loc=80,scale=15,size=100)
print(math_scores)
print(sci_scores)
```

```python
math_df=pd.DataFrame(math_scores)
sci_scores_df=pd.DataFrame(sci_scores)
print(math_df)
print(sci_scores_df)
```

```python
# prompt: Create a figure with two subplots:
# The first subplot should be an overlapping histogram showing the distribution of scores for both Math and Science. Use transparency (alpha) to make both distributions visible.
# The second subplot should be a scatter plot comparing math_scores (x-axis) against science_scores (y-axis) to see if there's a correlation

import matplotlib.pyplot as plt

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Subplot 1: Overlapping histograms
axes[0].hist(math_scores, bins=20, alpha=0.7, label='Math Scores')
axes[0].hist(sci_scores, bins=20, alpha=0.7, label='Science Scores')
axes[0].set_title('Distribution of Math and Science Scores')
axes[0].set_xlabel('Scores')
axes[0].set_ylabel('Frequency')
axes[0].legend()

# Subplot 2: Scatter plot
axes[1].scatter(math_scores, sci_scores, alpha=0.6)
axes[1].set_title('Math Scores vs. Science Scores')
axes[1].set_xlabel('Math Scores')
axes[1].set_ylabel('Science Scores')

plt.tight_layout()
plt.show()
```
