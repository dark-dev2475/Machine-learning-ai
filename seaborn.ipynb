data visuslaization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn has some built in datasets

```python
#total bill vs tip dataset
tips=sns.load_dataset('tips')
tips.head()
```

```python
#setting a theme for yourplot
sns.set_theme()
```

```python

#visulaize teh data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#oading the iris dataset
iris=sns.load_dataset('iris')
iris.head()
```

```python
#scatter plot
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
```

```python
#house price dtaset
from sklearn.datasets import fetch_california_housing
house_california=fetch_california_housing(as_frame=True)
house=house_california.frame
house.head()
```

```python
house=pd.DataFrame(house_california.data,columns=house_california.feature_names)
house['PRICE']=house_california.target
house.head()
```

distribution part

```python
sns.distplot(house['PRICE'])
```

correlation matrix
positive
negative

heat map

```python
correlation=house.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
